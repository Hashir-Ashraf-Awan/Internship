```python
import pandas as pd
import matplotlib.pyplot as plt
dataset=pd.read_csv("insurance_data.csv")
dataset.head()
```

```python
plt.scatter(dataset.age,dataset.bought_insurance,marker="*",color="blue")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 

x_train,x_test,y_train,y_test=train_test_split(dataset[['age']],dataset.bought_insurance,train_size=0.9)
```

```python
x_train,x_test
```

```python
model=LogisticRegression()
```

```python
model.fit(x_test,y_test)
```

```python
model.predict(x_test),x_test,y_test
```

```python
model.score(x_test,y_test)
```

```python
plt.scatter(x_test,y_test,marker="*",color="blue")
plt.plot()
```
