```python
import pandas as pd
import numpy as np

dataset=pd.read_csv("dataset.csv")
dataset.head

x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values

dataset.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # type: ignore
from sklearn.metrics import mean_squared_error
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
x_train,x_test,y_train,y_test
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
t=reg.predict(x_test)
t #predictedd
x_test
```

```python
y_test #actual values
```

```python
import matplotlib.pyplot as pl 
```

```python
pl.scatter(x_train,y_train, color="red")
pl.plot(x_train,reg.predict(x_train), color="blue")
pl.title("Linear Regression Salary vs Experiences")
pl.xlabel("Years of Experiece")
pl.ylabel("Salary")
pl.show()
```

```python
mse=mean_squared_error(y_test,t)
print("Mean Squared Error is ", mse)
```
