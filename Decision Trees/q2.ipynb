```python
import pandas as pd
dataset=pd.read_csv("titanic.csv")


sexes=dataset['Sex'].unique()
dataset.head()
dataset['Age']=dataset['Age'].fillna(value=dataset['Age'].mean())
dataset
```

```python
grouped_counts = dataset.groupby(['Sex', 'Survived']).size().unstack(fill_value=0)
counts_array = grouped_counts.values.flatten()
counts_array
sexes=['Women Survived','Women not Survived','Men not Survived','Men Survived']
import matplotlib.pyplot as plt
plt.pie(counts_array,labels=sexes, autopct='%1.1f%%', shadow=True,explode=[0,0,0.09,0],startangle=323)
plt.show()
```

```python

target=dataset['Survived']
mapping={'male':0, 'female':1}
dataset['Sex']=dataset['Sex'].map(mapping)
```

```python
input1 = dataset[['Sex', 'Pclass', 'Age', 'Fare']]
from sklearn import tree
tree=tree.DecisionTreeClassifier()
tree.fit(input1,target)
```

```python
tree.score(input1,target)
```

```python
tree.predict([['1','2','23','34']])
```
